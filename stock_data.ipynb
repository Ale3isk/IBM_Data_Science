# Extracting Stock Data Using Python Libraries

In this file I will extract financial data with the help of libraries:
- yfinance
- pandas

The I will extract information about the stock(s) and provide commentary. This is a simple Jupyter notebook with the aim of demonstrating how easily it is to extract financial data information using yfinance and pandas.

```python
#import the necessary libraries
import yfinance as yf
import pandas as pd
```

```python
#get the ticker object of the stock NVIDIA.
nvidia = yf.Ticker("NVDA")
print(nvidia)
```

```python
#use nvidia.info to get information about the stock. Returns dictionary.
nvidia.info.keys()
```

```python
#using a dictionary parsing we can extract keys and/or values of the dictionary
#extract stock information from the dictionary
for key in nvidia.info.keys():
    if key in ['quoteType', 'currency', 'shortName','symbol','country', 'sector']:
        print(nvidia.info[key])
```

```python
#Using the history() method in order to get the share price of the stock concerning a certain time interval.

# Here are some of the possible values for the period parameter and what they represent:

# period="1d": Download 1 day of historical data.
# period="5d": Download 5 days of historical data.
# period="1mo": Download 1 month of historical data.
# period="3mo": Download 3 months of historical data.
# period="6mo": Download 6 months of historical data.
# period="1y": Download 1 year of historical data.
# period="2y": Download 2 years of historical data.
# period="5y": Download 5 years of historical data.
# period="10y": Download 10 years of historical data.
# period="ytd": Download historical data since the beginning of the current year.
# period="max": Download all available historical data.


#extract the price of last day
nvidia.history(period="1 day")
```

```python
#extract the prices of last month
nvidia.history(period="1mo")
```

```python
#extracting the prices of last year
nvidia.history(period="1y").head()
```

```python
#we can pass the information onto a dataframe
df_nvidia = nvidia.history(period='max')
```

```python
#see the new dataframe
df_nvidia.head()
```

```python
#reset_index
df_nvidia.reset_index(inplace=True)
```

```python
#plot date vs closing price using Pandas
df_nvidia.plot(x='Date',y='Close', title='NVIDIA stock closing price - all time', ylabel='Price')
```

```python
#Extracting Dividends - returns a pandas series
nvidia.dividends
```

```python
#plot dividends
nvidia.dividends.plot()
```

```python
#Plot the volume traded
df_nvidia.plot('Date', 'Volume')
```
